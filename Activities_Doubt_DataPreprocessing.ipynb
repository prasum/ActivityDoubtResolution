```
import pandas as pd
import datetime as dt
```

```
act_df=pd.read_csv("activities.csv")
act_df.head(10)
```

Adding a user_type column in the activities data based on the activity type. This will help in identifying TA, User and System for analysis purposes.

```
def user_type(u):
  if(u['activity_type']=="assign"):
    return 'TA'
  if(u['activity_type']=="activate"):
    return 'TA'
  if(u['activity_type']=="reject"):
    return 'TA'
  if(u['activity_type']=="pending_action_required"):
    return 'TA'
  if(u['activity_type']=="pending_information_required"):
    return 'TA'
  if(u['activity_type']=="review_resolution"):
    return 'TA'
  if(u['activity_type']=="accept"):
    return 'TA'
  if(u['activity_type']=="reactivate"):
    return 'TA'
  if(u['activity_type']=="escalate"):
    return 'TA'
  if(u['activity_type']=="resolve"):
    return 'Student'
  if(u['activity_type']=="rate"):
    return 'Student'
  if(u['activity_type']=="review_not_resolved"):
    return 'Student'
  if(u['activity_type']=="not_pending_action_done"):
    return 'Student'
  if(u['activity_type']=="not_pending_info_submit"):
    return 'Student'
  if(u['activity_type']=="unassign"):
    return 'System'
  if(u['activity_type']=="inactivate"):
    return 'System'
  if(u['activity_type']=="pending_sms"):
    return 'System'
  if(u['activity_type']=="pending_reminder_first"):
    return 'System'
  if(u['activity_type']=="pending_reminder_second"):
    return 'System'
  if(u['activity_type']=="dead"):
    return 'System'
```

```
act_df['user_type']=act_df.apply(user_type,axis=1)
```

```
act_df.to_csv("activities_with_usertype.csv",index=False)
```

Getting the Data which has the activity type as Resolved.

```
act_df_resolved=act_df[act_df['activity_type']=="resolve"]
act_df_resolved.to_csv("activities_with_userresolve.csv",index=False)
```

Getting the data with user_type as TA

```
act_df_ta=act_df[act_df['user_type']=="TA"]
act_df_ta.to_csv("activities_with_ta.csv",index=False)
```

```
act_count_ta_df=act_df_ta.groupby("doubt_id")["user_id"].nunique().reset_index(name='TA_Count')
act_count_ta_df.to_csv("doubt_resolution_with_count_of_ta.csv",index=False)
```

Getting the TA who last worked on the doubt before it got resolved as the doubt assignment gets changed for the doubts.

```
act_doubt_ta=act_df_ta.groupby("doubt_id")["user_id"].nth(-1).reset_index(name='TA_ID')
act_doubt_ta.to_csv("doubt_resolution_ta.csv",index=False)
```

```
doubt_df=pd.read_csv("doubts.csv")
doubt_df.head(10)
```

Merging the doubt data and the user resolved data to get the Resolve timestamp

```
act_doubt_df = pd.merge(act_df_resolved, doubt_df, how='inner', left_on='doubt_id', right_on='id')
act_doubt_df.shape
```

Merging the doubt and the TA_Count data but here to prevent data loss right outer join is used as not all doubts have been resolved by the TAs. Some are user resolved.

```
act_doubt_df = pd.merge(act_count_ta_df, act_doubt_df, how='right', on='doubt_id')
act_doubt_df.shape
```

Merging the doubt and the corresponding TA ID data resolving the doubt. Here also to prevent the data loss right outer join is used as not all the doubts have been resolved by the TAs. Some are user resolved.

```
act_doubt_df = pd.merge(act_doubt_ta, act_doubt_df, how='right', on='doubt_id')
act_doubt_df.shape
```

Dropping the duplicate columns from the merged dataframe.

```
act_doubt_df=act_doubt_df.drop(['id_x', 'user_id_x','id_y','activity_type'], axis = 1) 
```

Renaming the columns of the merged dataframe due to conflicting column names and thus adding of suffixes to them.

```
act_doubt_df.rename(columns = {'created_at_x':'doubt_resolve_time', 'created_at_y':'doubt_creation_time', 
                              'user_id_y':'student_id'}, inplace = True)
act_doubt_df.head()
```

```
act_doubt_df.to_csv("act_doubt_df.csv",index=False)
```
